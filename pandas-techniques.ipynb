```python
import pandas as pd 
```

```python
#poetry dataset from kaggle
df=pd.read_csv('PoetryFoundationData.csv', index_col=0)
df.dropna()
```

```python
df['Poem'] = [str(x).replace("\r\r\n","") for x in df['Poem']]
df['Poet'] = [str(x).replace("\r\r\n","") for x in df['Poet']]
df['Title'] = [str(x).replace("\r\r\n","") for x in df['Title']]
df['Tags'] = [str(x).replace("\r\r\n","") for x in df['Tags']]
```

## Sorting DataFrame

```python
df.sort_values("Poet")
```

```python
df.sort_values(by="Poet",ascending=False)
```

```python
df.sort_values(by="Poet",ascending=False,kind="mergesort")
```

```python
df.sort_values(by=["Poem", "Poet"])
```

```python
df.sort_values(by=["Poet", "Poem", "Title"],ascending=[True, True, False])
```

```python
assigned_index_df = df.set_index(["Poet", "Poem"])
assigned_index_df.sort_index()
```

```python
df.sort_index(axis=1)
```
