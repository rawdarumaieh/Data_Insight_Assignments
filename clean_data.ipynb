```python

def range_perHour_salary(x):
    if pd.isnull(x):
        return np.nan
    elif ' /hour' in x:
        i=x.split('/hour')[0].replace('$',' ').strip()
        if '-' in i:
            mini=i.split('-')[0]
            maxi=i.split('-')[1]
            return mini, maxi
```

```python
df = df.assign(salary_perHour_Range=df['salary'].apply(range_perHour_salary)) 
```
